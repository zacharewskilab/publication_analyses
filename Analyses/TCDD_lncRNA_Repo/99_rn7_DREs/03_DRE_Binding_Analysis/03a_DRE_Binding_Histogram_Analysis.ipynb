# Visualization of rn7 DREs (histograms)

### 1. Import packages

```python
#Import necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### 2. Import DRE genomic coordinates

```python
#Import DRE data
DRE_Master = pd.read_csv('../02_CleanUp_DRE_BED_File/RN_Chromosomal_Location_of_DREs.txt', 
                     delimiter='\t', 
                     index_col=None)
DRE_Master.head(5)
```

```python
DRE_Master_Not_Na = DRE_Master[~DRE_Master['Dist_to_Start'].isna()]
DRE_Master_Not_Na
```

```python
DRE_in_lncRNAs = DRE_Master_Not_Na[DRE_Master_Not_Na['Gene_Biotype'] == 'lncRNA']
DRE_in_lncRNAs
```

```python
DRE_in_lncRNAs_liftover_only = DRE_in_lncRNAs[~DRE_in_lncRNAs['Mouse_ID'].isna()]
DRE_in_lncRNAs_liftover_only
```

```python
DRE_in_lncRNAs_rn7_only = DRE_in_lncRNAs[~DRE_in_lncRNAs['Rat_ID'].isna()]
DRE_in_lncRNAs_rn7_only
```

```python
DRE_in_proteins = DRE_Master_Not_Na[(DRE_Master_Not_Na['Gene_Biotype'] == 'mRNA')]# |(DRE_Master_Not_Na['Gene_Biotype'] == 'predicted_mRNA')]
DRE_in_proteins
```

```python
import os

directory = 'Figures'

# Check if the directory exists
if not os.path.exists(directory):
    # If it doesn't exist, create it
    os.makedirs(directory)
```

```python
DRE_in_lncRNAs_liftover_only['distance_kbp'] = DRE_in_lncRNAs_liftover_only['Dist_to_Start'] / 1000
DRE_in_lncRNAs_liftover_only
```

```python
# Convert distances to kilobase pairs
DRE_in_lncRNAs = DRE_in_lncRNAs.drop_duplicates(subset=['DRE_ID', 'Dist_to_Start'], keep='first')
DRE_in_lncRNAs['distance_kbp'] = DRE_in_lncRNAs['Dist_to_Start'] / 1000

# Filter distances to keep values between -10 and 10
filtered_distances = DRE_in_lncRNAs[(DRE_in_lncRNAs['distance_kbp'] >= -10) & (DRE_in_lncRNAs['distance_kbp'] <= 10)]['distance_kbp'].dropna()

# Plot histogram
plt.figure(figsize=(3.8, 1))
plt.hist(filtered_distances, bins=900)# , color='skyblue', edgecolor='black')
plt.xlabel('Distance to TSS (kb)')
plt.ylabel('Frequency')
plt.title('lncRNA Genomic Binding Location of DRE')
plt.xlim(-10, 10)
plt.ylim(0, 300)
plt.yticks([0, 150, 300])

# Save plot as PDF
plt.savefig('./Figures/DRE_Histogram_Small_lncRNA.pdf', format='pdf')

plt.show()
```
