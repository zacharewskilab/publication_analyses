# Visualization of rn7 AHR Enrichment Sites (histograms)

```python
#Import necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Import AhR Data
AHR_Master= pd.read_csv('../02_CleanUp_BED_File/rn7_Chromosomal_Location_of_AHR_Binding.txt', 
                            delimiter = '\t',
                            index_col=None)

AHR_Master
```

```python
AHR_Master['Transcript_Type'].unique()
```

```python
AHR_in_lncRNAs = AHR_Master[(AHR_Master['Transcript_Type'] == 'lncRNA')]
AHR_in_lncRNAs['Gene_ID'].nunique()
                    
```

```python
AHR_in_lncRNAs
```

```python
AHR_in_lncRNAs_liftover_only = AHR_in_lncRNAs[~AHR_in_lncRNAs['Mouse_ID'].isna()]
AHR_in_lncRNAs_liftover_only
```

```python
AHR_in_lncRNAs_rn7_only = AHR_in_lncRNAs[~AHR_in_lncRNAs['Rat_ID'].isna()]
AHR_in_lncRNAs_rn7_only
```

```python
AHR_in_proteins = AHR_Master[(AHR_Master['Transcript_Type'] == 'mRNA')]
AHR_in_proteins['Gene_ID'].nunique()
```

```python
AHR_in_proteins
```

```python
import os

directory = 'Figures'

# Check if the directory exists
if not os.path.exists(directory):
    # If it doesn't exist, create it
    os.makedirs(directory)
```

```python
# Convert distances to kilobase pairs
AHR_in_lncRNAs['distance_kbp'] = AHR_in_lncRNAs['Dist_to_Start'] / 1000

# Filter distances to keep values between -10 and 10
filtered_distances = AHR_in_lncRNAs[(AHR_in_lncRNAs['distance_kbp'] >= -10) & (AHR_in_lncRNAs['distance_kbp'] <= 10)]['distance_kbp'].dropna()

# Plot histogram
plt.figure(figsize=(3.8, 1))
plt.hist(filtered_distances, bins=300)# , color='skyblue', edgecolor='black')
plt.xlabel('Distance to TSS (kb)')
plt.ylabel('Frequency')
plt.title('lncRNA Genomic Binding Location of AHR')
plt.xlim(-10, 10)
plt.ylim(0,300)
plt.yticks([0, 150, 300])

# Save plot as PDF
plt.savefig('./Figures/AHR_Binding_Histogram_Small_lncRNA.pdf', format='pdf')

plt.show()
```
