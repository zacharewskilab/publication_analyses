# Process mm39 AHR enrichment sites

### 1. Import Packages

```python
import numpy as np
import pandas as pd
```

### 2. Import RAW data

```python
# Define the file path
file_path = "./RAW_Data/mm10_AHR_Binding_Regions_temp.bed"

# Read the file into a DataFrame
df = pd.read_csv(file_path, sep='\t', header=0, index_col=None)

# Display the DataFrame
df
```

```python
# Define the file path
file_path = "./RAW_Data/AHR_Enrichment_Male_2hr.txt"

# Read the file into a DataFrame
df2 = pd.read_csv(file_path, sep='\t', header=0, index_col=None)

df2_subset = df2[['Unique Peak ID', 'Fold-Change', 'FDR']]

# Display the DataFrame
df2_subset
```

```python
merged_df = df.merge(df2_subset, left_on="AHR_Site_Unique_ID", right_on="Unique Peak ID", how="left")
merged_df
```

```python
merged_df.to_csv('mm39_AHR_Binding_Regions.bed', sep='\t', index=False)
```
