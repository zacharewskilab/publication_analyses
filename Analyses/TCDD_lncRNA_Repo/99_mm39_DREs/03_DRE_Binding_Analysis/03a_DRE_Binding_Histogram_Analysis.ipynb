# Visualization of mm39 DREs (histograms)

### 1. Import packages

```python
#Import necessary libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### 2. Import DRE genomic coordinates

```python
#Import DRE data
DRE_Master = pd.read_csv('../02_CleanUp_DRE_BED_File/MMU_Chromosomal_Location_of_DREs.txt', 
                     delimiter='\t', 
                     index_col=None)
DRE_Master
```

```python
DRE_Master['Gene_Biotype'].unique()
```

```python
DRE_in_lncRNAs = DRE_Master[(DRE_Master['Gene_Biotype'] == 'lncRNA') |
                            (DRE_Master['Gene_Biotype'] == 'antisense') |
                            (DRE_Master['Gene_Biotype'] == 'lincRNA') |
                            (DRE_Master['Gene_Biotype'] == 'NR')| 
                            (DRE_Master['Gene_Biotype'] == 'lncOfInterest')]
DRE_in_lncRNAs['Gene_Name'].nunique()
```

```python
DRE_in_lncRNAs['Gene_Biotype'].unique()
```

```python
DRE_in_lncRNAs
```

```python
DRE_in_proteins = DRE_Master[(DRE_Master['Gene_Biotype'] == 'NM') |
                            (DRE_Master['Gene_Biotype'] == 'NM#NR')]
DRE_in_proteins['Gene_Name'].nunique()
```

```python
DRE_in_proteins
```

```python
import os

directory = 'Figures'

# Check if the directory exists
if not os.path.exists(directory):
    # If it doesn't exist, create it
    os.makedirs(directory)
```

```python
# Convert distances to kilobase pairs
DRE_in_lncRNAs['distance_kbp'] = DRE_in_lncRNAs['Dist_to_Start'] / 1000

# Filter distances to keep values between -10 and 10
filtered_distances = DRE_in_lncRNAs[(DRE_in_lncRNAs['distance_kbp'] >= -10) & (DRE_in_lncRNAs['distance_kbp'] <= 10)]['distance_kbp'].dropna()

# Plot histogram
plt.figure(figsize=(3.8, 1))
plt.hist(filtered_distances, bins=300)# , color='skyblue', edgecolor='black')
plt.xlabel('Distance to TSS (kb)')
plt.ylabel('Frequency')
plt.title('lncRNA Genomic Binding Location of DRE')
plt.xlim(-10, 10)
plt.ylim(0, 300)
plt.yticks([0, 150, 300])

# Save plot as PDF
plt.savefig('./Figures/DRE_Histogram_Small_lncRNA.pdf', format='pdf')

plt.show()
```

```python
# Convert distances to kilobase pairs
DRE_in_proteins['distance_kbp'] = DRE_in_proteins['Dist_to_Start'] / 1000

# Filter distances to keep values between -10 and 10
filtered_distances = DRE_in_proteins[(DRE_in_proteins['distance_kbp'] >= -10) & (DRE_in_proteins['distance_kbp'] <= 10)]['distance_kbp'].dropna()

# Plot histogram
plt.figure(figsize=(3.8, 1))
plt.hist(filtered_distances, bins=300)# , color='skyblue', edgecolor='black')
plt.xlabel('Distance to TSS (kb)')
plt.ylabel('Frequency')
plt.title('mRNA Genomic Binding Location of DRE')
plt.xlim(-10, 10)
plt.ylim(0, 300)
plt.yticks([0, 150, 300])

# Save plot as PDF
plt.savefig('./Figures/DRE_Histogram_Small_proteins.pdf', format='pdf')

plt.show()
```
